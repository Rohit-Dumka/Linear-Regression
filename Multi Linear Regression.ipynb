```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("homeprices_2.csv")
```

```python
df
```

```python
plt.scatter(df.area,df.price)
plt.show()
```

```python
plt.scatter(df.age,df.price)
plt.show()
#negative linear reg
```

```python
x=df[['area','bedrooms','age']]
x
```

```python
y=df.price
y
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()

# y=m1x1+ m2x2 + m3x3 + b
```

```python
model.fit(x,y)
```

```python
model.coef_
#          m1             m2              m3 (since -ve lin reg relationship  that's why -ve value)
```

```python
b=model.intercept_
b
```

```python
area=int(input("Enter the area of the house : "))
brooms = int(input("Enter the number of bed rooms : "))
age = int(input("Enter the age of the House : "))

price=model.predict([[area,brooms,age]])
print("Price of the house is :",price)
```

```python
model.score(x,y)
```
